```python
def reverse_a_string(given_string:str) -> str:
    """
    Function to reverse a string using for loop.
    Args:
        given_string(str): The string to be return.
    Returns:
        str: The reversed version of your string.
    Raises:
        TypeError: If input is not a string.
    """

    #validate input type

    if not isinstance(given_string,str):
        raise TypeError("Input must be a string")
    reverse_string = ""
    for i in given_string:
        reverse_string = i + reverse_string
    return reverse_string

    #main excution

if __name__ == "__main__":
    try:
     #taking input from the user.
     given_string = input("Enter the String : ")
     #calling the function and storing value in result;
     result = reverse_a_string(given_string)
     print(f"Reversed String : {result}")

       
    except TypeError as e:
        print(e)

    except Exception as e:
        print(e)
```

```python
def find_largest_number(number_list: list[float]) -> float:
    if not isinstance(number_list,list):
        raise TypeError("Input must be a list of numbers float or int")
    if not number_list:
        raise ValueError("list cannot be empty")
    for num in number_list:
        if not isinstance(num,(int,float)):
            raise TypeError("all elements must be numeric")
    largest = number_list[0]
    for num in number_list:
        if num>largest:
            largest = num
    return largest
if __name__ =="__main__":
    try:
        user_input = input("Enter numbers in the list int or float only: ")
        number_list = [float(item) for item in user_input.split()]
        result = find_largest_number(number_list)
        print("largest numbers in the list is :",result)

    except(TypeError,ValueError) as error:
        print(error)
    except Exception as e:
        print(e)
        
          
            
```
